¡Hola a todos, clase! Hoy vamos a aprender a dominar una de las herramientas más potentes de creación de video con inteligencia artificial: Google Veo (en sus versiones 2 y 3) a través de la plataforma Flow. Para ser directores expertos, el primer paso es entender la **composición**, que es cómo organizamos los elementos visuales para que el espectador vea exactamente lo que queremos.

Aquí tienen la guía completa y los ejercicios para que empiecen a crear como profesionales:

### 1. Formas de Componer (Tipos de Plano y Ángulo)
En Google Veo, la composición se reduce principalmente a la **posición de la cámara y la apertura del lente**. Estas son las formas que deben conocer:

*   **Close up (Acercamiento):** Para tomas donde la cámara está muy cerca del sujeto.
*   **Extremely close up (Primerísimo primer plano):** Para tomas aún más detalladas y cercanas.
*   **Wide shot (Plano general):** Para capturar una escena amplia y lejana.
*   **Frame (Encuadre):** La organización general de los elementos en el plano.
*   **Over the shoulder (Por encima del hombro):** Una toma clásica de conversación o perspectiva.
*   **Above the head (Cenital/Por encima de la cabeza):** Ángulo desde arriba del sujeto.
*   **Low angle (Ángulo bajo):** Cámara situada abajo mirando hacia arriba para dar importancia al sujeto.
*   **35 mm Film:** Palabra clave esencial si buscan un **aspecto cinematográfico** profesional.

### 2. Cómo darle indicaciones a la IA (La "Bioestructura")
Para que la IA no ignore sus órdenes, el orden de los factores **sí altera el producto**. Deben seguir esta estructura exacta:

1.  **Tipo de toma:** Incluyan aquí la **composición y los movimientos de cámara** (ej. Dolly, Pan, Zoom).
2.  **Sujeto + Acción + Detalles:** Describan quién es el protagonista y qué está haciendo.
3.  **Descripción del ambiente:** Aquí va el contexto general del escenario.
4.  **Resultado final:** Indiquen cómo concluye la escena o si hay una transición.
5.  **Visuales clave:** Elementos específicos que **deben** aparecer en el cuadro.

**El Secreto de Experto:** Para que la IA realice la indicación tal cual, **escriban solamente lo que está en su frame o toma**. Si piden un *close up* del rostro pero luego describen los zapatos y pantalones del personaje, la IA se confundirá y omitirá el primer plano para mostrar la ropa. **Sean asertivos:** si no se ve en cámara, no lo describan.

### 3. Ejercicios Prácticos: "Tomándole la mano a Veo"

Siguiendo el orden del video, realicen estos ejercicios (recuerden que los prompts deben ser en **inglés** para mejores resultados):

*   **Ejercicio 1: Composición Básica y Estilo.**
    *   *Tarea:* Creen una escena usando la palabra clave **"35 mm Film"** y un **"Close up"**.
    *   *Objetivo:* Lograr una toma cinematográfica de un rostro sin describir nada del cuerpo inferior para asegurar que la cámara se mantenga cerca.

*   **Ejercicio 2: Movimiento y Acción.**
    *   *Tarea:* Usen la estructura: `Tipo de toma (Tracking shot) + Sujeto (Astronauta) + Acción (Caminando)`.
    *   *Objetivo:* Practicar cómo la cámara sigue al sujeto en movimiento de forma fluida.

*   **Ejercicio 3: Contexto y Mood (Atmósfera).**
    *   *Tarea:* Generen un video que incluya el **Contexto Temporal** (hora del día) y un **Mood** (ej. "Golden hour light" o "Atmósfera tensa").
    *   *Objetivo:* Aprender a diferenciar un video genérico de uno profesional mediante la iluminación y el entorno.

*   **Ejercicio 4: Consistencia con Imagen a Video.**
    *   *Tarea:* Suban una imagen propia a Flow, seleccionen **"Fotogramas a video"** y denle una instrucción de movimiento simple como **"Zoom in"**.
    *   *Objetivo:* Mantener la consistencia del personaje usando una referencia visual en lugar de solo texto.

**Nota importante para la clase:** Google Veo tiene un límite ideal de entre **200 a 250 palabras** por prompt. No describan de más; si saturan el texto, la IA empezará a omitir detalles o creará cambios de toma extraños. ¡A practicar!

Para que la inteligencia artificial de Google Veo (específicamente en su versión 2 y 3) genere resultados óptimos, el orden de los factores **sí altera el producto**, ya que el motor tiende a dar mayor importancia a la parte inicial del texto.

Siguiendo lo que se describe en las fuentes como la **"bioestructura"**, el orden específico que deben seguir tus estudiantes para construir un prompt es el siguiente:

1.  **Tipo de toma:** Es el primer elemento y debe incluir tanto la **composición** (posición de la cámara y apertura del lente) como los **movimientos de cámara** (ej. tracking shots, dolly, pan).
2.  **Sujeto + Acción + Detalles:** En esta sección se describe quién es el protagonista, qué actividad está realizando (acciones como caminar, saltar o nadar) y detalles específicos que le den contexto a dichas acciones.
3.  **Descripción del ambiente:** Aquí se debe incluir el **contexto general del escenario**, que abarca la ubicación, la hora del día y las condiciones climáticas.
4.  **Resultado final:** En este punto se indica cómo concluye la escena o si existe alguna transición específica en la cámara o cambio en la ambientación.
5.  **Visuales clave:** Se deben listar los elementos específicos que son fundamentales que aparezcan dentro del cuadro.

### Recomendaciones estratégicas para la estructura:
*   **Prioridad inicial:** Debido a que la IA prioriza el inicio del prompt, una estructura incorrecta puede causar que se pierdan detalles importantes al final de la instrucción.
*   **Asertividad en el encuadre:** Es vital escribir **solamente lo que está en el frame** o toma actual. Si se pide un primer plano (*close up*) pero luego se describen los zapatos del sujeto, la IA se confundirá y omitirá la instrucción de cercanía para mostrar la vestimenta completa.
*   **Idioma:** Aunque se pueden incluir diálogos en otros idiomas para BO3, las instrucciones de la estructura deben estar en **inglés**, ya que es el único lenguaje optimizado para que el motor capte adecuadamente las indicaciones.
*   **Extensión ideal:** Para mantener la coherencia y evitar que la IA omita animaciones o cree cambios extraños, el prompt debe tener entre **200 a 250 palabras**.
