A continuación se presenta una **mini-secuencia coherente de 5 clips**, diseñada para funcionar tanto de forma independiente como encadenada en una progresión narrativa clara, siguiendo estrictamente la estructura y duración solicitadas. Cada clip dura **exactamente 3 segundos**.

---

## Clip 1: **Resonancia Inicial**

**Duración:** 3 segundos

**Narrativa temporal:**

* **0.0s – 0.8s:** Dragon Pol Z aparece de espaldas, ligeramente inclinado hacia adelante. El escenario es un cielo nublado con rocas flotando lentamente. El cabello está relajado, apenas moviéndose por el viento.
* **0.9s – 1.8s:** Un pulso de Ki recorre su cuerpo desde los pies hasta la cabeza. El cabello comienza a erizarse de forma sutil; pequeñas partículas de luz azul emergen del suelo.
* **1.9s – 3.0s:** Gira la cabeza hacia cámara en un perfil decidido. Un aura azul tenue se enciende alrededor de su silueta y permanece estable.

**VFX:** Aura azul translúcida de baja intensidad. Partículas ascendentes suaves. Distorsión ligera del aire alrededor del cuerpo.

**Sonido sugerido:** Zumbido grave y continuo de energía acumulándose.

---

## Clip 2: **Pulso de Preparación**

**Duración:** 3 segundos

**Narrativa temporal:**

* **0.0s – 0.7s:** Plano medio frontal. Dragon Pol Z flexiona los brazos y baja el centro de gravedad. El aura azul se vuelve más densa.
* **0.8s – 1.6s:** El cabello se eleva con mayor fuerza; aparecen destellos eléctricos cian entre los mechones. El suelo bajo sus pies vibra.
* **1.7s – 3.0s:** El aura se expande brevemente como una onda circular y luego se contrae, quedando comprimida alrededor del torso.

**VFX:** Aura azul con bordes blancos. Chispas eléctricas intermitentes. Onda expansiva semitransparente sobre el suelo.

**Sonido sugerido:** “Power-up hum” con pulsos rítmicos.

---

## Clip 3: **Núcleo de Ataque**

**Duración:** 3 segundos

**Narrativa temporal:**

* **0.0s – 0.6s:** Dragon Pol Z separa los pies y extiende ambos brazos hacia adelante, manos abiertas.
* **0.7s – 1.5s:** Entre sus palmas se forma una esfera de Ki azul-dorado, girando sobre sí misma. El rostro muestra tensión y enfoque absoluto.
* **1.6s – 3.0s:** La esfera crece y se estabiliza, iluminando el rostro y el pecho. El aura general se intensifica pero permanece contenida.

**VFX:** Esfera energética con capas concéntricas, partículas orbitando y destellos internos. Iluminación dinámica sobre el personaje.

**Sonido sugerido:** Sonido agudo de energía condensándose, con vibración creciente.

---

## Clip 4: **Descarga Instantánea**

**Duración:** 3 segundos

**Narrativa temporal:**

* **0.0s – 0.4s:** Breve pausa tensa. La esfera de Ki alcanza su máximo brillo.
* **0.5s – 1.2s:** Dragon Pol Z empuja violentamente las manos hacia adelante. El ataque se libera en forma de haz concentrado.
* **1.3s – 3.0s:** El rayo atraviesa la escena y sale de cuadro. El retroceso mueve el cabello y el aura hacia atrás; el entorno se sobreexpone brevemente.

**VFX:** Haz de energía azul-blanco con núcleo brillante. Motion blur direccional. Flash luminoso al momento del disparo.

**Sonido sugerido:** “Energy beam blast” potente con cola reverberante.

---

## Clip 5: **Dominio Final**

**Duración:** 3 segundos

**Narrativa temporal:**

* **0.0s – 0.9s:** El rayo se disipa fuera de cámara. Polvo y partículas flotan en el aire frente a Dragon Pol Z.
* **1.0s – 1.9s:** El aura disminuye gradualmente. El cabello desciende, aún con leves chispas residuales.
* **2.0s – 3.0s:** Pose final victoriosa: de pie, brazos relajados, mirada firme hacia cámara. Fondo despejado con cielo luminoso.

**VFX:** Partículas residuales de Ki disipándose. Aura residual débil. Ligero vapor energético alrededor del cuerpo.

**Sonido sugerido:** Eco lejano del impacto seguido de un silencio controlado.

La secuencia completa narra una progresión clara: **activación → preparación → concentración → liberación → resolución**, alineada con la estética Shonen tipo Dragon Ball Z y optimizada para clips ultracortos de alto impacto visual, listos para animación o generación con herramientas de video por IA.
