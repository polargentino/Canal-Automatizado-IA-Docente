¡Excelente elección! Usar un formato **JSON** es la forma más profesional de comunicarte con LLMs como Grok o Qwen para mantener la consistencia, ya que les permite procesar la información como datos estructurados y no solo como texto plano.

Para lograr que la IA mantenga la consistencia del personaje (Lucas) y la atmósfera, he diseñado una estructura que incluye un **"bloque de contexto global"** (que define quién es Lucas y el estilo visual) y luego la **lista de escenas**.

Copia y pega el siguiente bloque completo en tu chat con Grok o Qwen.

### Guía para usar esto en Kdenlive:

Una vez que generes los clips con Grok/Qwen (usando los prompts del campo `"video_generation_prompt"` de cada escena) y tengas los audios (usando el campo `"voice_over_text"`), sigue este flujo en Kdenlive:

1.  **Pista V1 (Video):** Coloca los clips del 1 al 5 en orden.
    *   *Tip:* Usa transiciones suaves (Disolver) entre la escena 1 y 2. Usa cortes duros (cortes secos) entre la 4 y la 5 para aumentar el impacto del susto.
2.  **Pista A1 (Voz):** Coloca tus archivos de voz en off. Asegúrate de dejar "aire" (silencio) entre frases para crear tensión.
3.  **Pista A2 (Ambiente):** Descarga un sonido de "Room Tone" (sonido de habitación vacía) y ponlo de fondo en todo el video a volumen bajo (-15dB o -20dB).
4.  **Pista A3 (SFX):** Coloca los efectos de sonido específicos (el latido del corazón, el zumbido eléctrico, el abrir del libro) sincronizados exactamente con la acción visual.
5.  **Corrección de Color (Opcional):** En Kdenlive, aplica el efecto **"Curvas de Bézier"** o **"Niveles"** para oscurecer un poco más las sombras y asegurar que todos los clips tengan el mismo tono oscuro y amarillento de la lámpara.
