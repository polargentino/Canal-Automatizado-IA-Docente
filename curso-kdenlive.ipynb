¡Hola! Con gusto te extraigo la información sobre la presentación de este curso de Kdenlive.

Este video es el **Tema 0: Presentación del curso** de Kdenlive, un editor de video libre, gratuito y multiplataforma, presentado por Juan Febles del canal Podcast Linux.

El objetivo principal del curso es enseñar la base del programa para realizar producciones audiovisuales, eliminar la barrera inicial y facilitar un flujo de trabajo que ahorre tiempo y potencie la creatividad [[00:41](http://www.youtube.com/watch?v=Np8CP66G2Pg&t=41)].

El presentador detalla los 16 temas iniciales (además de la presentación), organizados en tres bloques principales:

### **Bloque 1: Introducción y Flujo de Trabajo (Temas 1-6)**

Este bloque se centra en la preparación y las herramientas básicas del programa:

* **Descarga, Instalación y Documentación:** La importancia de la documentación para no perderse [[01:17](http://www.youtube.com/watch?v=Np8CP66G2Pg&t=77)].
* **Organización, Áreas y Preferencias:** Cómo organizar los diferentes paneles y áreas de trabajo y ajustar las preferencias a tus necesidades [[01:25](http://www.youtube.com/watch?v=Np8CP66G2Pg&t=85)].
* **Proyectos y Clips:** Creación de un proyecto y cómo implementar clips (vídeos, audios, imágenes) [[01:35](http://www.youtube.com/watch?v=Np8CP66G2Pg&t=95)].
* **La Línea de Tiempo:** La parte más importante del editor, donde se colocan los archivos para crear secuencias [[01:58](http://www.youtube.com/watch?v=Np8CP66G2Pg&t=118)].
* **Exportar el Vídeo:** Conseguir el producto final, aprovechando al máximo los diferentes tipos de archivos y contenedores [[02:16](http://www.youtube.com/watch?v=Np8CP66G2Pg&t=136)].
* **Atajos del Teclado:** Herramientas para ganar fluidez y trabajar en menos tiempo [[02:31](http://www.youtube.com/watch?v=Np8CP66G2Pg&t=151)].

### **Bloque 2: Elementos Clave y Efectos (Temas 7-12)**

Este bloque cubre técnicas y efectos para mejorar las producciones:

* **Transiciones:** Cómo realizar el paso suave de un clip a otro para evitar cortes bruscos [[03:01](http://www.youtube.com/watch?v=Np8CP66G2Pg&t=181)].
* **Efectos de Vídeo:** Cómo transformar y añadir dinamismo a tus vídeos [[03:20](http://www.youtube.com/watch?v=Np8CP66G2Pg&t=200)].
* **Efectos de Audio:** Cómo sacar un mayor partido al sonido, una parte crucial de cualquier video [[03:35](http://www.youtube.com/watch?v=Np8CP66G2Pg&t=215)].
* **Corrección de Color:** Dar la intensidad y luminosidad adecuada para transmitir lo que se desea [[03:50](http://www.youtube.com/watch?v=Np8CP66G2Pg&t=230)].
* **Fotogramas Clave:** Usar puntos en el tiempo para que los efectos no sean homogéneos y crear movimientos dinámicos [[04:10](http://www.youtube.com/watch?v=Np8CP66G2Pg&t=250)].
* **Añadir Títulos:** Un soporte importante para agregar información a los vídeos [[04:25](http://www.youtube.com/watch?v=Np8CP66G2Pg&t=265)].

### **Bloque 3: Extras y Ejemplos Prácticos (Temas 13-16)**

Este bloque finaliza con ejemplos prácticos para aplicar lo aprendido:

* **Crear una Intro:** Un ejemplo sencillo de cómo crear una introducción [[04:40](http://www.youtube.com/watch?v=Np8CP66G2Pg&t=280)].
* **Vídeos de Fotos:** Cómo crear carruseles de fotos rápidamente [[04:54](http://www.youtube.com/watch?v=Np8CP66G2Pg&t=294)].
* **Voz en Off:** Cómo grabar la voz en *off* directamente en Kdenlive [[05:01](http://www.youtube.com/watch?v=Np8CP66G2Pg&t=301)].
* **Sincronizar Sonidos:** Cómo sincronizar el audio de la cámara con el de una grabación externa (como un micrófono de solapa) [[05:16](http://www.youtube.com/watch?v=Np8CP66G2Pg&t=316)].

El curso se publicará tanto en YouTube como en PeerTube, y todo el contenido está bajo licencia **Creative Commons Attribution Compartir Igual** [[05:59](http://www.youtube.com/watch?v=Np8CP66G2Pg&t=359)].

# Número 2:
¡Muy bien! Continuamos con el curso. Aquí tienes el resumen del tercer video.

Este video corresponde al **Tema 2: Organización, áreas y preferencias** de Kdenlive, donde se explica cómo adaptar la interfaz del programa a tus necesidades.

### **1. Organización y Áreas Personalizables [[00:30](http://www.youtube.com/watch?v=hQW6_msk5TE&t=30)]**

La clave de este tema es que la estructura inicial de Kdenlive (con vistas como Registro, Edición, Audio, Efectos y Color) es totalmente adaptable.

* **Adaptación de Paneles:** Puedes mover, reestructurar o eliminar cualquier panel (como el monitor de clip o la carpeta de proyectos) utilizando los **tiradores** para ajustarlos al tamaño y posición que prefieras [[02:19](http://www.youtube.com/watch?v=hQW6_msk5TE&t=139)].
* **Guardar la Organización:** Una vez que hayas modificado el diseño a tu gusto, es crucial que guardes tu nueva organización de área yendo a la opción de menú **Ver > Guardar organización** [[02:38](http://www.youtube.com/watch?v=hQW6_msk5TE&t=158)].
* **Importar y Exportar:** Puedes **exportar** tu configuración de área (por ejemplo, la vista "Edición") para llevarla a otros ordenadores o usarla como copia de seguridad, y también puedes **importarla** [[03:09](http://www.youtube.com/watch?v=hQW6_msk5TE&t=189)].
* **Vistas Predeterminadas:** Kdenlive viene con vistas predefinidas (`Registro`, `Edición`, `Audio`, `Efectos`, etc.), cada una orientada a un flujo de trabajo específico. Se recomienda ser metódico y seguir un flujo de trabajo constante [[04:24](http://www.youtube.com/watch?v=hQW6_msk5TE&t=264)].

### **2. Preferencias de Kdenlive [[04:55](http://www.youtube.com/watch?v=hQW6_msk5TE&t=295)]**

Desde el menú de preferencias, puedes configurar aspectos visuales y de rendimiento del programa:

* **Temas de Color:** Puedes cambiar el tema de colores (por omisión, el presentador usa el tema oscuro) [[05:04](http://www.youtube.com/watch?v=hQW6_msk5TE&t=304)].
* **Iconos:** Se puede elegir el paquete de iconos, como **Brisa**, para una mejor visualización [[05:12](http://www.youtube.com/watch?v=hQW6_msk5TE&t=312)].
* **Configuración General (Configurar Kdenlive):** Permite acceder a varias configuraciones avanzadas, incluyendo:
    * **Opciones Predeterminadas de Proyecto:** (Se tratará en el Tema 3) [[05:34](http://www.youtube.com/watch?v=hQW6_msk5TE&t=334)].
    * **Clips de Reemplazo (Proxy Clips):** Una opción muy importante si tu ordenador es un poco antiguo o no tiene un rendimiento óptimo para la edición de vídeo, ya que mejora la fluidez del trabajo [[05:43](http://www.youtube.com/watch?v=hQW6_msk5TE&t=343)].
