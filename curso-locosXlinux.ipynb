### Resumen del Contenido del Video (Curso Kdenlive #1)

Este video es la primera parte de un tutorial sobre el uso del editor de video libre y gratuito **Kdenlive**. El creador utiliza grabaciones de *skate* de su amigo Martín como ejemplo [[01:03](http://www.youtube.com/watch?v=oaKLkccCsp8&t=63)].

Los puntos principales cubiertos en el tutorial son:

* **Interfaz y Configuración:** El presentador muestra la interfaz de Kdenlive (versión 20.08.3 en ese momento) [[01:32](http://www.youtube.com/watch?v=oaKLkccCsp8&t=92)] y explica cómo ajustar el tema y las opciones predeterminadas del proyecto, que él establece en 1080p a 30 FPS [[02:44](http://www.youtube.com/watch?v=oaKLkccCsp8&t=164)].
* **Importar y Organizar Clips:** Muestra cómo arrastrar clips a la sección de Proyecto [[02:07](http://www.youtube.com/watch?v=oaKLkccCsp8&t=127)] y cómo organizarlos en la línea de tiempo, diferenciando entre el monitor de clips y el monitor de proyecto [[03:57](http://www.youtube.com/watch?v=oaKLkccCsp8&t=237)].
* **Recortes (Cortes):** Enseña el atajo de teclado `Shift + R` para cortar clips rápidamente en la línea de tiempo [[07:49](http://www.youtube.com/watch?v=oaKLkccCsp8&t=469)] y eliminar las partes sobrantes.
* **Transiciones Simples:** Explica cómo crear una transición de "barrido" (desvanecimiento) colocando dos clips superpuestos en pistas diferentes de video y audio [[10:49](http://www.youtube.com/watch?v=oaKLkccCsp8&t=649)].
* **Efectos de Audio y Video:** Muestra cómo añadir un efecto de *Fade Out* (desvanecimiento de salida) y *Fade In* (desvanecimiento de entrada) al audio y al video para evitar cambios bruscos [[12:35](http://www.youtube.com/watch?v=oaKLkccCsp8&t=755)] y [[14:52](http://www.youtube.com/watch?v=oaKLkccCsp8&t=892)].
* **Música de Fondo:** Muestra cómo agregar una pista musical y cómo ajustar su volumen o duración [[13:44](http://www.youtube.com/watch?v=oaKLkccCsp8&t=824)].
* **Procesar (Renderizar) el Video:** Finalmente, explica el proceso de renderizado, usando el formato MP4 con el códec H.264, y la importancia de guardar el proyecto para futuras ediciones [[16:47](http://www.youtube.com/watch?v=oaKLkccCsp8&t=1007)].

El creador concluye preguntando a la audiencia si quieren más partes del curso de Kdenlive o tutoriales sobre **GIMP** o **Audacity** [[18:22](http://www.youtube.com/watch?v=oaKLkccCsp8&t=1102)].
