Basado en el tutorial de **AulaMusical**, aquí tienes los puntos más importantes para dominar Audacity desde cero:

### 1. Configuración Inicial y Dispositivos [[03:55](http://www.youtube.com/watch?v=FMSbMAK0llI&t=235)]

* **Servidor de audio:** Elige entre MME (más compatible), Direct Sound (baja latencia) o WASAPI (mejor calidad, ideal para grabar lo que suena en el PC).
* **Dispositivos:** Selecciona correctamente tu micrófono (entrada) y tus altavoces o auriculares (salida) en el botón de **Configuración de audio**.
* **Canales:** Usa **Mono** para voces (un solo micrófono) y **Estéreo** para música o instrumentos complejos.

### 2. Control de Niveles y Grabación [[07:38](http://www.youtube.com/watch?v=FMSbMAK0llI&t=458)]

* **Monitorización:** Haz clic en el icono del micrófono para activar la monitorización silenciosa y ver los niveles antes de grabar.
* **Nivel ideal:** Intenta que tu voz fluctúe entre **-24 y -20 dB** para evitar distorsiones y tener margen de edición.
* **Regla de 3 segundos:** Deja 3 segundos de silencio antes de hablar para capturar el "ruido de fondo" de la habitación, lo que facilitará su eliminación posterior.

### 3. Navegación y Edición Básica [[13:24](http://www.youtube.com/watch?v=FMSbMAK0llI&t=804)]

* **Atajos esenciales:** * `Barra espaciadora`: Reproducir/Detener.
* `Tecla P`: Pausa.
* `F1`: Herramienta de selección.
* `Ctrl + 1` / `Ctrl + 3`: Zoom acercar/alejar.


* **Selección:** Puedes resaltar partes del audio y usar la tecla `Suprimir` para eliminar errores o silencios innecesarios.
* **Importar:** Arrastra archivos de música directamente desde tu carpeta a Audacity para añadirlos como pistas nuevas [[19:27](http://www.youtube.com/watch?v=FMSbMAK0llI&t=1167)].

### 4. Mejora de la Calidad del Audio (Efectos) [[27:38](http://www.youtube.com/watch?v=FMSbMAK0llI&t=1658)]

* **Normalizar:** Ajusta el volumen general a un nivel uniforme.
* **Reducción de Ruido:** Selecciona el silencio inicial (perfil de ruido) y luego aplícalo a toda la pista para limpiar zumbidos de fondo [[29:13](http://www.youtube.com/watch?v=FMSbMAK0llI&t=1753)].
* **Compresor:** Ayuda a que las partes bajas de tu voz suenen más cerca de las partes altas, dando consistencia.
* **Ecualización:** Para voces masculinas, se recomienda realzar graves ligeros y claridad en los medios; para femeninas, filtrar frecuencias muy bajas para evitar opacidad [[34:06](http://www.youtube.com/watch?v=FMSbMAK0llI&t=2046)].

### 5. Mezcla Profesional (Voz y Música) [[38:57](http://www.youtube.com/watch?v=FMSbMAK0llI&t=2337)]

* **Herramienta Envolvente:** Permite bajar manualmente el volumen de la música cuando tú hablas.
* **Auto Duck:** Es un proceso automático que baja el volumen de la música "mágicamente" cada vez que detecta tu voz en la pista de abajo [[40:06](http://www.youtube.com/watch?v=FMSbMAK0llI&t=2406)].
* **Fade In / Fade Out:** Aplica desvanecimientos al inicio y final de la música para transiciones suaves [[41:36](http://www.youtube.com/watch?v=FMSbMAK0llI&t=2496)].

### 6. Guardar vs. Exportar [[20:58](http://www.youtube.com/watch?v=FMSbMAK0llI&t=1258)]

* **Guardar Proyecto (.aup3):** Guarda todas las pistas por separado para seguir editando después.
* **Exportar (MP3/WAV):** Crea el archivo final que puedes enviar por WhatsApp, correo o subir a plataformas.
